```python
import numpy as np
import pandas as pd
from top2vec import Top2Vec

model = Top2Vec.load("ESGnie_emissions_model_V1")

data = pd.read_csv('ESGnie_target_descriptions.csv')
data.drop_duplicates(inplace=True)

docs = list(set(data['target_explanation_english'].tolist()))
```

```python
topic_sizes, topic_nums = model.get_topic_sizes()
```

```python
topic_sizes, topic_nums = model.get_topic_sizes()
classified_docs = {}
for i in range(75):
    documents, document_scores, document_ids = model.search_documents_by_topic(topic_num=topic_nums[i], num_docs=topic_sizes[i])
    document_dict = dict(zip(document_ids, documents))
    classified_docs.update(document_dict)
```

```python
classified_docs
```

```python
#create dataframe with classified_docs with document id as index
df = pd.DataFrame(data=classified_docs.values(), index=list(classified_docs.keys()))
```

```python
df.reset_index(inplace=True)
```

```python
df.rename(columns={'index':'document_id',0:'Document'}, inplace=True)
```

```python
def get_topic_num(x):
    return model.get_documents_topics(doc_ids=[x])[0][0]

def get_topic_socre(x):
    return model.get_documents_topics(doc_ids=[x])[1][0]
```

```python
df['topic_num'] = df['document_id'].apply(get_topic_num)
```

```python
df['topic_score'] = df['document_id'].apply(get_topic_socre)
```
